```python
import pandas as pd
import numpy as np

#upload data
df = pd.read_csv(r"D:\Courses\Data Analysis\NTI\Project\WA_Fn-UseC_-Telco-Customer-Churn.csv")

#Detect null values
print(df.isnull().sum())
```

```python

#استبدال الفراغات بقيمة مفقودة 
#استخدمنا np.nan لي ؟ عشانها float  
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)

#تعبئة القيم الفارغة
df['TotalCharges'] = df['TotalCharges'].fillna(0.0)
print("\n--- 2. معالجة 11 قيمة Null في TotalCharges بـ 0.0 ---")

# tenure مدة الخدمة يمثل عدد الشهور وهو رقم صحيحintger 
df['tenure'] = df['tenure'].astype(int)

# MonthlyCharges الرسوم الشهرية رقم مالي يحتوي على كسور عشرية
# نستخدم float للحفاظ على الدقة العشرية
df['MonthlyCharges'] = df['MonthlyCharges'].astype(float)

#TotalCharges إجمالي الرسوم رقم مالي يحتوي على كسور عشرية
#نستخدم float للحفاظ على الدقة العشرية
df['TotalCharges'] = df['TotalCharges'].astype(float)
print(df.dtypes['TotalCharges'])
```

```python

#  عمل عمود جديد tenure_category
# تعريف الدالة التي تقسم 'tenure' إلى فئات
def categorize_tenure(tenure):
    if 1 <= tenure <= 10:
        return 'Short'
    elif 11 <= tenure <= 50:
        return 'Long'
    elif tenure > 50:
        return 'Very Long'
    else:                   
        return 'Zero Tenure'
```

```python
# تطبيق الدالة لإنشاء العمود
df['tenure_category'] = df['tenure'].apply(categorize_tenure)

#  تبديل القيم في عمود SeniorCitizen
df['SeniorCitizen'] = df['SeniorCitizen'].replace({
    0: 'Younger',
    1: 'Older'
})
```
